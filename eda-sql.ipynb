```python
import sqlite3
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"

df = pd.read_csv(url)

df.head()
```

```python
conn = sqlite3.connect(':memory:')

df.to_sql(
    "TIPS",
    conn,
    index=False,
    if_exists='replace'
)
```

```python
query = """
SELECT day,
COUNT(*) as Total
FROM TIPS
GROUP BY day
"""

pd.read_sql(query, conn)
```

```python
query2 = """
SELECT sex,
AVG(total_bill) as AverageBill
FROM TIPS
GROUP BY sex
"""

pd.read_sql(query2, conn)
```
