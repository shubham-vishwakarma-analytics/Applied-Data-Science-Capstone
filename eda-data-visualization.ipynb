```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset("tips")

df.head()
```

```python
sns.scatterplot(
    x='total_bill',
    y='tip',
    hue='sex',
    data=df
)

plt.show()
```

```python
sns.boxplot(
    x='day',
    y='total_bill',
    data=df
)

plt.show()
```

```python
df.groupby('day')['tip'].mean().plot(kind='bar')

plt.ylabel("Average Tip")

plt.show()
```
